### Model Description

[MiDaS](https://arxiv.org/abs/1907.01341) computes relative inverse depth from a single image. The repository provides multiple models that cover different use cases ranging from a small, high-speed model to a very large model that provide the highest accuracy. The models have been trained on 10 distinct datasets using
multi-objective optimization to ensure high quality on a wide range of inputs.

### Example Usage

Download an image from the PyTorch homepage

```python
import cv2
import torch
import urllib.request

import matplotlib.pyplot as plt

url, filename = ("https://github.com/pytorch/hub/raw/master/images/dog.jpg", "dog.jpg")
urllib.request.urlretrieve(url, filename)
```

Load a model (see [https://github.com/intel-isl/MiDaS/#Accuracy](https://github.com/intel-isl/MiDaS/#Accuracy) for an overview)

```python
model_type = "DPT_Large"     # MiDaS v3 - Large     (highest accuracy, slowest inference speed)
#model_type = "DPT_Hybrid"   # MiDaS v3 - Hybrid    (medium accuracy, medium inference speed)
#model_type = "MiDaS_small"  # MiDaS v2.1 - Small   (lowest accuracy, highest inference speed)

midas = torch.hub.load("intel-isl/MiDaS", model_type)
```

```python
import torch
import numpy as np

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
midas.to(device)
midas.eval()

dummy_input = torch.randn(1, 3, 384, 384, dtype=torch.float32).to(device)

with torch.no_grad():
    output = midas(dummy_input)

print("Output shape:", output.shape)
```

```python
import cv2
import numpy as np

midas_transforms = torch.hub.load("intel-isl/MiDaS", "transforms")
transform = midas_transforms.dpt_transform

image_path = "2d1.jpg"

img = cv2.imread(image_path)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

input_batch = transform(img).to(device)

with torch.no_grad():
    prediction = midas(input_batch)

prediction = torch.nn.functional.interpolate(
    prediction.unsqueeze(1),
    size=img.shape[:2],
    mode="bicubic",
    align_corners=False,
).squeeze(0).squeeze(0)

depth_map = prediction.cpu().numpy()
print("Depth map shape:", depth_map.shape)
print("Depth min:", depth_map.min(), "max:", depth_map.max())
```

Move model to GPU if available

```python
device = torch.device("cuda") if torch.cuda.is_available() else torch.device("cpu")
midas.to(device)
midas.eval()
```

Load transforms to resize and normalize the image for large or small model

```python
midas_transforms = torch.hub.load("intel-isl/MiDaS", "transforms")

if model_type == "DPT_Large" or model_type == "DPT_Hybrid":
    transform = midas_transforms.dpt_transform
else:
    transform = midas_transforms.small_transform
```

Load image and apply transforms

```python
img = cv2.imread(filename)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

input_batch = transform(img).to(device)
```

Predict and resize to original resolution

```python
with torch.no_grad():
    prediction = midas(input_batch)

    prediction = torch.nn.functional.interpolate(
        prediction.unsqueeze(1),
        size=img.shape[:2],
        mode="bicubic",
        align_corners=False,
    ).squeeze()

output = prediction.cpu().numpy()
```

Show result

```python
import numpy as np
import matplotlib.pyplot as plt

def run_midas_on_image(path):
    img = cv2.imread(path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    input_batch = transform(img).to(device)
    with torch.no_grad():
        prediction = midas(input_batch)
        prediction = torch.nn.functional.interpolate(
            prediction.unsqueeze(1),
            size=img.shape[:2],
            mode="bicubic",
            align_corners=False,
        ).squeeze()
    depth = prediction.cpu().numpy()
    return img, depth

test_images = [
    "dog.jpg",
    "2d1.jpg"

]

for p in test_images:
    img, depth = run_midas_on_image(p)
    plt.figure(figsize=(8, 4))
    plt.subplot(1, 2, 1)
    plt.imshow(img)
    plt.axis("off")
    plt.subplot(1, 2, 2)
    plt.imshow(depth)
    plt.axis("off")
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def generate_stereo_pair(img, depth, max_shift=80):
    h, w, _ = img.shape
    d = depth.astype(np.float32)
    d_min = d.min()
    d_max = d.max()
    d = (d - d_min) / (d_max - d_min + 1e-8)
    disparity = (0.5 - d) * max_shift
    x, y = np.meshgrid(
        np.arange(w, dtype=np.float32),
        np.arange(h, dtype=np.float32)
    )
    map_x_left = x + disparity
    map_x_right = x - disparity
    map_y = y.astype(np.float32)
    left = cv2.remap(img, map_x_left, map_y, interpolation=cv2.INTER_LINEAR, borderMode=cv2.BORDER_CONSTANT)
    right = cv2.remap(img, map_x_right, map_y, interpolation=cv2.INTER_LINEAR, borderMode=cv2.BORDER_CONSTANT)
    margin = int(max_shift)
    if margin * 2 < w:
        left = left[:, margin:-margin, :]
        right = right[:, margin:-margin, :]
    stereo_pair = np.concatenate([left, right], axis=1)
    return left, right, stereo_pair

def make_anaglyph(left, right):
    left_gray = cv2.cvtColor(left, cv2.COLOR_RGB2GRAY)
    right_gray = cv2.cvtColor(right, cv2.COLOR_RGB2GRAY)
    anaglyph = np.zeros_like(left)
    anaglyph[..., 0] = left_gray
    anaglyph[..., 1] = right_gray
    anaglyph[..., 2] = right_gray
    return anaglyph

for p in test_images:
    img, depth = run_midas_on_image(p)
    left, right, stereo_pair = generate_stereo_pair(img, depth, max_shift=80)
    anaglyph = make_anaglyph(left, right)
    plt.figure(figsize=(18, 4))
    plt.subplot(1, 4, 1)
    plt.imshow(img)
    plt.axis("off")
    plt.subplot(1, 4, 2)
    plt.imshow(depth)
    plt.axis("off")
    plt.subplot(1, 4, 3)
    plt.imshow(stereo_pair)
    plt.axis("off")
    plt.subplot(1, 4, 4)
    plt.imshow(anaglyph)
    plt.axis("off")
    plt.show()
```
